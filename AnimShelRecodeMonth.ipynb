# recode movementDate1 to reflect month names

```python
shelterMonthName.movementDate1.value_counts()
```

```python
def movementDate1R (series):
    if series == 1:
        return "January"
    if series == 2:
        return "February"
    if series == 3:
        return "March"
    if series == 4:
        return "April"
    if series == 5:
        return "May"
    if series == 6:
        return "June"
    if series == 7:
        return "July"
    if series == 8:
        return "August"
    if series == 9:
        return "September"
    if series == 10:
        return "October"
    if series == 11:
        return "November"
    if series == 12:
        return "December"
shelterMonthName["movementDate1R"]= shelterMonthName["movementDate1"].apply(movementDate1R)
```
